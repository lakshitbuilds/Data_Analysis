```python
import pandas as pd
df = pd.read_csv("Mobile_Dataset.csv")
```

```python
# 1. Show first 5 rows
df.head(5)
# 2. Show last 10 rows
df.tail(10)
# 3. Find total rows and columns
df.shape
# 4. Show all column names
df.columns
# 5. Check datatype of each column
df.dtypes
# 6. Find missing values
df.isnull().sum()
# 7. Find duplicate rows
df[df.duplicated()]
# 8. Count unique companies
df["Company Name"].unique()
```

```python
df.columns
```

```python
# Find all phones launched in 2023.
df[df["Launched Year"]==2023]
```

```python
# Find all Samsung phones.
df[df["Company Name"] == "Samsung"]
```

```python
# Find all Xiaomi phones.
df[df["Company Name"] == "Xiaomi"]
```

```python
df["Battery Capacity"] = df["Battery Capacity"].str.replace("mAh", "")
df["Battery Capacity"] =df["Battery Capacity"].str.replace(",", "").astype(int)
```

```python
# Find phones with battery greater than 5000 mAh.
df[df["Battery Capacity"]>5000]
```
