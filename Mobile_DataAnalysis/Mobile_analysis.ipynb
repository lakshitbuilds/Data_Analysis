```python
# Create a new column called `"Price Category"`:

# * Budget
# * Mid-Range
# * Premium
df["Price Category"] = df[
    "Launched Price (India)"
].apply(
    lambda x:
    "Budget" if x < 20000
    else "Mid-Range" if x < 50000
    else "Premium"
)
```

```python
# Find percentage of premium phones.
total = df["Price Category"].value_counts().sum()

premium = df["Price Category"].value_counts()["Premium"]

percentage = (premium / total) * 100

percentage
```
